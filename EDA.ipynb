```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3)
fig.set_figheight(4)
fig.set_figwidth(15)

df['Body Material'].value_counts().plot.bar(title="Body Materials",ax=ax1)
df['Neck Material'].value_counts().plot.bar(title="Neck Materials",ax=ax2)
df['Fretboard Material'].value_counts().plot.bar(title="Fretboard Materials",ax=ax3)
```

```python
most_used_woods_body = list(df['Body Material'].value_counts()[:10].index)
most_used_woods_neck = list(df['Neck Material'].value_counts()[:2].index)
most_used_woods_fret = list(df['Fretboard Material'].value_counts()[:10].index)
```

```python
plt.figure(figsize=(12,10))
plt.xticks(rotation=90)
sns.boxplot(x="Body Material", y="Price", data=df[df['Body Material'].isin(most_used_woods_body)])
```

```python
plt.figure(figsize=(12,10))
plt.xticks(rotation=90)
sns.boxplot(x="Neck Material", y="Price", data=df[df['Neck Material'].isin(most_used_woods_neck)])
```
